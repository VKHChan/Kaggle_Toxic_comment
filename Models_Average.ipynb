### Using weighted average of models

```python
import numpy as np
import pandas as pd
```

```python
# Bidirectional LSTM baseline model
y_lstm = pd.read_csv("Results/baseline_LSTM.csv")
```

```python
# LR with Tfidf
y_LR = pd.read_csv("Results/toxic_lr.csv")
```

```python
target_col = ['toxic', 'severe_toxic', 'obscene', 'threat','insult', 'identity_hate']
```

```python
y_avg = y_lstm.copy()
```

```python
# Model1 (0.5, 0.5)
# Model2 (0.4, 0.6)
w_lstm = 0.4
w_LR = 0.6
```

```python
y_avg[target_col] = w_lstm*y_lstm[target_col] + w_LR*y_LR[target_col]
```
